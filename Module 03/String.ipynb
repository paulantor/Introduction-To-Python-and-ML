### String Methode

```python
message=""" Wellcome to String Methode
here we learn many methode"""
# STRING lENGTH 
print(len(message))
# FIND A SUBSTRING IN A MAIN STRING
print("learn" in message)
print("Paul" in message)
# CONVERT LOWER LATTER  and UPPER
low_message=message.lower()
up_message=message.upper()
print(low_message)
print(up_message)
# KON INDEX THEKE SUBSTRING SURU HOISE ER JONNO DORKER .find
find_string=message.find("Methode")
print(find_string)
print(message.rfind("methode"))
# SUBSTRING KOTOBER ASE ER JONNO .count
print(low_message.count("methode"))
# REPLACE A SUBSTRING WITH ANOTHER SUBSTRING MAIN STRING KINTU REPLACE HOI NA
new_string=message.replace("methode","empty")
print(new_string)
```
