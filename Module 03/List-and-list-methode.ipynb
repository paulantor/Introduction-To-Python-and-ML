### LIST 

```python
#List sclicing
arr1=[12,223,12,34,455,665]
arr2=arr1[1:4]
print(arr2)
```

```python
#Adding an item in a list use insert or append insert a [position,value]

arr1.append(10000)
print(arr1)
arr1.insert(6,200000)
print(arr1)

```

```python
#Deleting 
# last index value delete
arr1.pop()
print(arr1)
# speacific value delete
arr1.remove(200000)
print(arr1)
```

### SORTING

```python
arr1.sort()
print(arr1)
arr1.reverse()
print(arr1)
arr1.sort(reverse=True)
print(arr1)
```
