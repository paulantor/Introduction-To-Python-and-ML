### Tuple


```python
tup=(10,39,45,"str",True)
print(tup,type(tup))
print(tup[3])
print(tup[0:3])
```

```python
# Tuple methode count,index
tup=(10,39,45,"str",True,10)
print(tup.count(10))
print(tup.index(45))
```

### SET

```python
s={1,2,3,4,56,8}
if(3 in s):
    print("YES")
else:
    print("NO")
```

```python
# ADD a value in set
s={1,2,3,4,56,8}
s.add(12)
s.pop()
s.remove(56)
print(s)

```

```python
a={1,3,4,5}
b={3,7,8,0}
print(a.union(b))
print(a.issubset(b))
```

### Dictionary 

```python
# {KEY,VALUE}
dic={'Name':'Antor','Age':'23','ID':'CE22044'}
print(dic["Name"])
print(dic.get("Age"))
```

```python
# Adding a value
dic["Num"]=30
print(dic)
dic.update({"Num":0})
print(dic)
```

```python
# DELETE
del dic["mob"]
print(dic)
```
