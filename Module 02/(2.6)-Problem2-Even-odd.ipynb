```python
n=int(input())
num=input().split()
even=0; odd=0; positive=0; negative=0
for i in range(n):
    p=int(num[i])
    if(p%2!=0):
        odd+=1
    else:
        even+=1
    if(p>0):
        positive+=1
    elif(p<0):
        negative+=1
print(f"Even: { even}")                
print(f"Odd: { odd}")                
print(f"Positive: { positive}")                
print(f"Negative: { negative}")                
```
