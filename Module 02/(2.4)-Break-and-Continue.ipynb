```python
accuracy=1
for i in range(1,100):
    if(accuracy%2!=0):
        print(f"{i} The Matchine Learning model is Precising")
        accuracy+=1
        continue
    else:
        if(accuracy==98  ):
            print(f"{i} The Matchine Learning model is Excellent")
            break
        else:
            print(f"{i} The Matchine Learning model is Good")    
    accuracy+=1        
```
