```python
val1="ODD";val2="EVEN"
for i in range(1,100,1):
    if((i%2)==0):
        print(f"{i} is {val2}")
    else:
        print(f"{i} is {val1}")    

# Here Range are contain three parameters. First one is starting point, second one is ending point and third one is step. By default step is 1. So if we do not provide step then it will be 1. So in this case we can also write for i in range(1,100) instead of for i in range(1,100,1)
# f" is used for string formatting. It allows us to embed expressions inside string literals, using curly braces {}. The expressions inside the curly braces are evaluated at runtime and their values are inserted into the string at the corresponding positions. In this case, {i} will be replaced with the current value of i, and {val1} or {val2} will be replaced with the corresponding string "ODD" or "EVEN".        
```
