```python
#Python a input default string a thake eta ke type cast kore amra onno type e convert korte pari. Jemon int, float etc.
age =int (input("Enter your age: ")) 
height = float (input("Enter your height in meters: "))
name= input("Enter your name: ")
print(age,type(age ))
print(height,type(height))
print(name,type(name))
```

```python
#oparator 
x=100;y=50
add=x+y
sub=x-y
mul=x*y
div=x/y
rem=x%y
print(add,sub,mul,div,rem)
```

```python
#comparison oparator always boolean value return kore. Jemon True, False
print(x>=y) #greater than or equal
print(x<=y) #  less than or equal
print(x==y) # equal to
print(x!=y) # not equal to
```

```python
#logical oparator AND ,OR ,NOT
a=100;b=50;c=90
result1=(a>b) and (c>b) # AND oparator duto condition true holei true return kore
result2=(a>b) and (c<b)
result3=(a<b) or (c>b) # OR opartor duto condition er modhye jekono ekta true holei true return kore
result4=not (a>b)

print(result1,result2,result3,result4)
```
